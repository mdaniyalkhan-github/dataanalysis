```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="ticks", palette="rocket")
```

```python
#reading the file
df = pd.read_csv("jewelry.csv", header = None)
```

```python
df.head()
```

```python
#providing new column names as the dataset didnt had proper column names
df.columns = [
    "Timestamp", "OrderID", "PurchasedID", "Quantity", "CategoryID", 
    "Jewelry_Type", "BrandID", "Price", "UserID", 
    "Gender", "Color", "Material", "Gemstone"
]
```

```python
#converting and splitting the Timestamp Column for more deep analysis
df['Timestamp'] = pd.to_datetime(df['Timestamp'], utc=True)
df['date'] = df['Timestamp'].dt.day
df['yearsold'] = df['Timestamp'].dt.year
df['month'] = df['Timestamp'].dt.month
```

```python
df.head()
```

```python
#dropping the null values in CategoryID column
df = df.dropna(subset=['CategoryID'])
df.info()
```

```python
#converting months to quarters for more insights 
def categorize_quarter(month):
    if month in [1, 2, 3]:
        return 'Q1'
    elif month in [4, 5, 6]:
        return 'Q2'
    elif month in [7, 8, 9]:
        return 'Q3'
    elif month in [10, 11, 12]:
        return 'Q4'
    else:
        return 'Unknown' 
  
```

```python
df_copy['Quarter'] = df_copy['month'].apply(categorize_quarter) #using apply feature to merge the data
```

```python
#since the dataset didnt have enough data for the year 2018, decided to drop the rows that included 2018 
df_copy = df_copy[df_copy['yearsold'] != 2018]
```

```python
#dropping the duplicates turns out the data had many duplicated values
df_copy.drop_duplicates(inplace = True)
```

```python
#for the corr matrix included only numerical values
numeric = df_copy.select_dtypes(include=[np.number])
corr_matrix = numeric.corr()
```

```python
#corr matrix
sns.heatmap(corr_matrix, cmap='coolwarm', vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#total number of orders each year
plt.figure(figsize=(10, 6))
order_counts = df_copy['yearsold'].value_counts().sort_index().reset_index().sort_values(by= 'count', ascending = False)
sns.barplot(x="yearsold", y="count", hue = 'yearsold' ,data=order_counts)
plt.xlabel('Years')
plt.ylabel('No. Of Orders')
plt.title('Number Of Orders Per Year')
plt.legend(title='Years')
plt.grid(True)
plt.show()
```

```python
#total number of orders by each type
plt.figure(figsize=(15, 6))
order_counts = df_copy['Jewelry_Type'].value_counts().sort_index().reset_index().sort_values(by= 'count', ascending = False).head(8)
sns.barplot(x="Jewelry_Type", y="count", hue = 'Jewelry_Type' ,data=order_counts)
plt.xlabel('Jewelry Type')
plt.ylabel('No. Of Orders')
plt.title('Number Of Orders By Jewelry Tyoe')
plt.grid(True)
plt.show()
```

```python
order_counts = df_copy.groupby(['Jewelry_Type', 'yearsold']).size().reset_index(name='count')

top_jewelry_types = order_counts.groupby('Jewelry_Type')['count'].sum().nlargest(8).index
order_counts = order_counts[order_counts['Jewelry_Type'].isin(top_jewelry_types)].sort_values(by = 'count', ascending = False)

plt.figure(figsize=(15, 6))
sns.barplot(x="Jewelry_Type", y="count", hue="yearsold", data=order_counts)

plt.xlabel('Jewelry Type')
plt.ylabel('No of Orders')
plt.title('Top 8 Jewelry Types by Orders and Year')
plt.legend(title='Years')
plt.grid(True)
plt.show()
```

```python
order_counts = df_copy.groupby(['yearsold', 'Jewelry_Type']).size().reset_index(name='count')

top_jewelry_types = order_counts.groupby('yearsold')['count'].sum().nlargest(10).index
order_counts = order_counts[order_counts['yearsold'].isin(top_jewelry_types)].sort_values(by = 'count', ascending = False)

plt.figure(figsize=(15, 6))
sns.barplot(x="yearsold", y="count", hue="Jewelry_Type", data=order_counts)

plt.xlabel('Years')
plt.ylabel('No. Of Orders')
plt.title('No. Of Orders By Year and Jewelry Type')
plt.legend(title='Jewelry Type')
plt.grid(True)
plt.show()
```
